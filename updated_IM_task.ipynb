```python
all_sheets = pd.read_excel('Instructor Manager Task Feedback Sample.xlsx', sheet_name = None)
```

```python
overall = all_sheets['Overall Class Feedback']
la = all_sheets['Instructor A Feedback']
lb = all_sheets['Instructor B Feedback']
ia = all_sheets['Instructional Associate Feedbac']
```

```python
column_name2 = "Is there anything you would like to share with us about the content?"

if column_name2 in overall.columns:
    overall['Polarity'] = overall[column_name2].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
    overall['Subjectivity'] = overall[column_name2].apply(lambda x: TextBlob(str(x)).sentiment.subjectivity)

    print(overall[[column_name2, 'Polarity', 'Subjectivity']])
else:
    print(f"Column '{column_name2}' does not exist in the DataFrame.")
```
